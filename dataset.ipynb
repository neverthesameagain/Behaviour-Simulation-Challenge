```python
df = pd.read_excel("/content/drive/MyDrive/interiit/content_simulation_train(1).xlsx")
```

```python
img_text = pd.concat([df['media'], df['content']], axis=1)
```

```python
content_list = []
image_ids = []
```

```python
image_dir = '/content/dataset/images'
os.makedirs(image_dir,exist_ok=True)
```

```python
def download_image(url, filepath):
  response = requests.get(url)
  with open(filepath, 'wb') as f:
    f.write(response.content)
```

```python
#changeeeee
index = 63600
img_text = img_text
```

```python
for index1, row in img_text.iterrows():
    media_str = row['media']
    if 'Photo' in media_str:
        preview_start = media_str.find("previewUrl='") + len("previewUrl='")
        preview_end = media_str.find("'", preview_start)
        preview_url = media_str[preview_start:preview_end]

        full_start = media_str.find("fullUrl='") + len("fullUrl='")
        full_end = media_str.find("'", full_start)
        full_url = media_str[full_start:full_end]

        image_id = f"photo_{index}"
        image_path = os.path.join(image_dir, f"{image_id}.jpg")
        download_image(full_url, image_path)

        image_ids.append(image_id)
        content_list.append(row['content'])


    elif 'Video' in media_str:
        thumbnail_start = media_str.find("thumbnailUrl='") + len("thumbnailUrl='")
        thumbnail_end = media_str.find("'", thumbnail_start)
        thumbnail_url = media_str[thumbnail_start:thumbnail_end]

        image_id = f"video_{index}"
        image_path = os.path.join(image_dir, f"{image_id}.jpg")
        download_image(thumbnail_url, image_path)

        image_ids.append(image_id)
        content_list.append(row['content'])
    index+=1
    print(index)
```
